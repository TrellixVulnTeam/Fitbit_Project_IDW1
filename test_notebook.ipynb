```python
import pandas as pd
import os
import json
import datetime as dict
```

```python
path = r'/Users/ericcollins/Documents/Python/Fitbit_Project/MyFitbitData/EricCollins/Sleep/'
df_sleep = pd.DataFrame()
for file in os.listdir(path):
    filepath = (path+'/'+file)
    if file.startswith('sleep') == True & file.endswith('.json') == True:
        # load data using Python JSON module
        with open(filepath,'r') as f:
            data = json.loads(f.read())
        # Flatten data
            df_nested_list = pd.json_normalize(data)
            df_sleep = pd.concat([df_sleep,df_nested_list])

df_sleep = df_sleep[['dateOfSleep','startTime','endTime','duration','minutesAsleep','minutesAwake','timeInBed','levels.summary.restless.minutes','levels.summary.deep.minutes','levels.summary.light.minutes','levels.summary.rem.minutes']]
df_sleep=df_sleep.rename(columns={'dateOfSleep':'date','levels.summary.restless.minutes':'restlessMinutes','levels.summary.deep.minutes':'minutesDeep',
                        'levels.summary.light.minutes':'minutesLight','levels.summary.rem.minutes':'minutesREM'})
df_sleep['date']=pd.to_datetime(df_sleep['date'])

```

```python
path = r'/Users/ericcollins/Documents/Python/Fitbit_Project/MyFitbitData/EricCollins/Physical Activity/'
df_zoneMinutes = pd.DataFrame()
for file in os.listdir(path):
    filepath = (path+'/'+file)
    if file.startswith('Active Zone Minutes') & file.endswith('.csv'):
        data = pd.read_csv(filepath)
        df_zoneMinutes = pd.concat([df_zoneMinutes,data])
df_zoneMinutes = df_zoneMinutes.sort_values(by='date_time')
import datetime as dt
df_zoneMinutes['date_time']=pd.to_datetime(df_zoneMinutes['date_time'])
df_zoneMinutes['date'] = df_zoneMinutes['date_time'].dt.date
df_zoneMinutes['date'] = pd.to_datetime(df_zoneMinutes['date'])
df_zoneMinutes['time'] = df_zoneMinutes['date_time'].dt.time
df_zoneMinutes = df_zoneMinutes.groupby(by='date',as_index = False).sum()
df_zoneMinutes = df_zoneMinutes.rename(columns={'total_minutes':'total_zone_minutes'})
```

```python
df = pd.merge(df_sleep,df_zoneMinutes,on='date')
```
